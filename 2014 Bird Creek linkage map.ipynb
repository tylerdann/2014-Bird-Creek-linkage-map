# Background on rounds of sequencing

Two families were chosen for sequencing, offspring from female 101 and female 110.  Both gynogenetic diploids and haploids were included in library prep and sequencing with the following sample sizes: 
101: haploid=49, gyno diploid=23
107: haploid=118
Together with dams 101 and 102 we prepared two libraries of 96 individuals each.

Libraries were mistakenly swapped on the instrument during the first round of sequencing, resulting in incorrect titrations for the second round of sequencing and inbalanced coverage among individuals.  A third round of sequencing was conducted but didn't result in adequate coverage for the smaller group of 101 haploid individuals to be useful in estimating recombination frequencies.  Finally, we used the instrument training run on ADF&G's NextSeq 500 to try and resurrect 101 haploid individuals and incorporate variation from dam 101 on the linkage map.

## Description of the four rounds of sequencing for 2014 Bird Creek haploids

Protocol when the two 2014 Bird Creek libraries (PHap1 and PHap2) were created was to conduct two rounds of sequencing to equalize coverage among individuals, titrating before the second round according to retained reads from the first round.

Unfortunately the two libraries were swapped when sequenced at the Core lab for the first round, resulting in incorrect titrations and no equalizing of coverage.  A third round was done on a subset of individuals to rescue them from too few reads.  The final round (unpacked above) was acquired during ADF&G's training run.
